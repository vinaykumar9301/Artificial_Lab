```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/sample_data/mnist_train_small.csv')
```

```python
data
```

```python
X_train = data.copy()
y_train = X_train.pop('6')
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
test = pd.read_csv('/content/sample_data/mnist_test.csv')
test
```

```python
X_test = test.copy()
y_test = X_test.pop('7')
```

```python
import tensorflow as tf
from tensorflow import keras

ann = keras.models.Sequential([
    keras.layers.Dense(units=128, activation='relu', input_shape=(X_train.shape[1],)),
    keras.layers.Dense(units=10, activation='softmax')  # 10 units for 10 classes
])
ann.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
ann.fit(X_train, y_train, batch_size=32, epochs=100)
```

```python
# Assuming you have X_test and y_test loaded

# Evaluate the model on the test data
test_loss, test_accuracy = ann.evaluate(X_test, y_test)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```
